```python
#Plotting graphs
import matplotlib.pyplot as plt
import numpy as np
from math import*
x=np.arange(-6,2*pi,0.1)
s=np.sin(x)
c=np.cos(x)

plt.plot(x,s,label="sinx")
plt.plot(x,c,label="cosx")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#from math import*
x=np.arange(-6,2*3.14,0.1)
s=np.sin(x)**2
c=np.cos(x)**2
plt.plot(x,s,label="sin^2(x)")
plt.plot(x,c,label="cos^2(x)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import*
x=np.linspace(-10,10,50)
y=np.exp(-x)

plt.plot(x,np.exp(x),label='e^x')
plt.plot(x,y,label='e^-x')
plt.xlabel("x------->")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import*
import cmath
x=np.arange(-6,6,0.1)
y=np.cos(x)+np.sin(x)
z=np.sin(x)**2
plt.plot(x,y,label='cosx+sinx')
plt.plot(x,z,label='sin^2x')
plt.legend()
plt.show()
```
