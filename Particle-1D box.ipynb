## $Plotting\space \Psi\space and\space \Psi*\Psi \space of \space a \space particle \space in \space a \space 1D \space potential \space well: $

```python
import matplotlib.pyplot as plt
import numpy as np


#Values for a and x
x_list = np.linspace(0,1,100)
a = 1

def psi(n,a,x):
    return np.sqrt(2/a)*np.sin(n*np.pi*x/a)

def psi_2(n,a,x):
    return np.square(psi(n,a,x))

plt.figure(figsize=(15,10))


for n in range(1,4):
#Empty lists for energy and psi wave
    psi_2_list = []
    psi_list = []
    for x in x_list:
        psi_2_list.append(psi_2(n,a,x))
        psi_list.append(psi(n,a,x))
       
    #Plotting Psi vs x
    plt.subplot(3,2,n)
    plt.plot(x_list, psi_list,label=r'$\Psi$')
    plt.xlabel("a")
    #plt.ylabel(r'$\Psi$')
    plt.legend()
    plt.title("n="+str(n))
    plt.grid()
    
    #Plotting Psi*Psi vs x    
    plt.subplot(3,2,n)
    plt.plot(x_list, psi_2_list,label=r'$\Psi*\Psi$')
    plt.xlabel("a")
    #plt.ylabel(r'$\Psi*\Psi$')
    plt.legend()
    plt.title("n="+str(n))
    plt.grid()
    
    
    
```
